```python
import requests
import re
import pandas as pd
from bs4 import BeautifulSoup
```

```python
r = requests.get("https://en.wikipedia.org/wiki/Fifth-generation_fighter")
r
```

```python
soup = BeautifulSoup(r.content,"html.parser")
soup
```

```python
td_element = soup.find_all("td")
td_element
```

```python
list_fighter =[]
for td in td_element:
    text = td.get_text()
    list_fighter.append(text)

list_fighter
```

```python
search_first_row_text = "USA"
search_last_row_text = "240 km"
```

```python
search_index_first_row=list_fighter.index(search_first_row_text)
search_index_last_row=list_fighter.index(search_last_row_text)

print("The index of the'USA' in the above list is: ",search_index_first_row )
print("The index of the'240 km' in the above list is: ",search_index_last_row)
```

```python
new_list=[]
for i in range(search_index_first_row+0,search_index_last_row+6):
    new_list.append(list_fighter[i])
print(new_list)
```

```python
F22=[]
for i in range(0,len(new_list),6):
    F22.append(new_list[i])
print(F22)
```

```python
F35A=[]
for i in range(1,len(new_list),6):
    F35A.append(new_list[i])
print(F35A)
```

```python
F35B=[]
for i in range(2,len(new_list),6):
    F35B.append(new_list[i])
print(F35B)
```

```python
F35C=[]
for i in range(3,len(new_list),6):
    F35C.append(new_list[i])
print(F35C)
```

```python
Su57=[]
for i in range(4,len(new_list),6):
    Su57.append(new_list[i])
print(Su57)
```

```python
J20=[]
for i in range(5,len(new_list),6):
    J20.append(new_list[i])
print(J20)
```

```python
final_dict = {"F-22":F22, "F-35A":F35A, "F-35B":F35B, "F-35C":F35C, "Su-57":Su57, "J-20":J20}
final_dict
```

```python
df = pd.DataFrame(final_dict)
df
```
